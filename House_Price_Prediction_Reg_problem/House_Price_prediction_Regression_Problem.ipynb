# Drop unnecessary columns

```python
X = train.drop(labels=["Id","SalePrice"],axis=1)
Y = train[["SalePrice"]]
```

# Remove Outliers

```python
from PM77 import standardize,outliers
ss = standardize(X)
ol = outliers(ss)
```

```python
X = X.drop(index=ol,axis=0)
Y = Y.drop(index=ol,axis=0)
```

```python
Y.shape
```

```python
X = X.reset_index(drop=True)
Y = Y.reset_index(drop=True)
```

# Preprocessing


```python
from PM77 import preprocessing
Xnew = preprocessing(X)
```

```python
Xnew.head()
```

```python
Xnew.isna().sum()
```

# train test split

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(Xnew,Y,test_size=0.2,random_state=21)
```

# OLS model

```python
from statsmodels.api import OLS, add_constant
ols = OLS(ytrain, add_constant(xtrain))
model = ols.fit()
model.summary()
col_to_drop = model.pvalues.sort_values().index[-1]
xtrain = xtrain.drop(labels=col_to_drop,axis=1)
xtest = xtest.drop(labels=col_to_drop,axis=1)
rsq = model.rsquared_adj
print("Dropped: column:",col_to_drop,"\tRsquared:",round(rsq,4))
```

```python
for i in range(0,10):
    from statsmodels.api import OLS, add_constant
    ols = OLS(ytrain, add_constant(xtrain))
    model = ols.fit()
    model.summary()
    col_to_drop = model.pvalues.sort_values().index[-1]
    xtrain = xtrain.drop(labels=col_to_drop,axis=1)
    xtest = xtest.drop(labels=col_to_drop,axis=1)
    rsq = model.rsquared_adj
    print("Dropped: column:",col_to_drop,"\tRsquared:",round(rsq,4))
```

# Create a Linear Regression model on the basis of selected columns

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(xtrain,ytrain)
tr_pred = model.predict(xtrain)
ts_pred = model.predict(xtest)
from sklearn.metrics import mean_absolute_error
tr_err = mean_absolute_error(ytrain,tr_pred)
ts_err = mean_absolute_error(ytest,ts_pred)
tr_err,ts_err
```

# Regularization

```python
from sklearn.linear_model import Ridge
rr = Ridge(alpha=0.090)
model = rr.fit(xtrain,ytrain)
tr_pred = model.predict(xtrain)
ts_pred = model.predict(xtest)
from sklearn.metrics import mean_absolute_error
tr_err = mean_absolute_error(ytrain,tr_pred)
ts_err = mean_absolute_error(ytest,ts_pred)
tr_err,ts_err
```

```python
alphas = []
q = 0.001
for i in range(1000):
    alphas.append(q)
    q = q+1
```

```python
tr = []
ts = []
for i in alphas:
    from sklearn.linear_model import Ridge
    rr = Ridge(alpha=i)
    model = rr.fit(xtrain,ytrain)
    tr_pred = model.predict(xtrain)
    ts_pred = model.predict(xtest)
    from sklearn.metrics import mean_absolute_error
    tr_err = mean_absolute_error(ytrain,tr_pred)
    ts_err = mean_absolute_error(ytest,ts_pred)
    tr.append(tr_err)
    ts.append(ts_err)
```

```python
import matplotlib.pyplot as plt
plt.plot(tr,c="red")
plt.plot(ts,c="blue")
```
